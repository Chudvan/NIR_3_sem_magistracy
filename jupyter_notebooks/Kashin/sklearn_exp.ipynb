```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
```

# work with all data

```python
data_dir = '/home/ivan/Рабочий стол/Учёба/Git/NIR_3_sem_magistracy/Data'
data = pd.read_csv(os.path.join(data_dir, 'data.csv'), sep = '\t')
data
```

```python
data.iloc[:, 2:11]
```

```python
X = data[['Neutral', 'Happy', 'Sad', 'Angry', 'Surprised', 'Scared', 'Disgusted']]
y = data[['Valence', 'Arousal']]
```

split dataframe

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

create models

```python
def ridge_regression(X_train, y_train, alpha_value):
    ridge_model = Ridge(alpha=alpha_value)
    ridge_model.fit(X_train, y_train)
    return ridge_model

def lasso_regression(X_train, y_train, alpha_value):
    lasso_model = Lasso(alpha=alpha_value)
    lasso_model.fit(X_train, y_train)
    return lasso_model

def elastic_net_regression(X_train, y_train, alpha_value, l1_ratio_value):
    elastic_net_model = ElasticNet(alpha=alpha_value, l1_ratio=l1_ratio_value)
    elastic_net_model.fit(X_train, y_train)
    return elastic_net_model

# для полиноминальной регрессии используем Ridge регресиию, но можно и другие потестить
def polynomial_regression(X_train, y_train, degree):
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    model.fit(X_train, y_train)
    return model
```

Расчет метрики качества, можно заменить на другое


```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mae = {}
    for idx, col in enumerate(y_test.columns):
      mae[col] = mean_absolute_error(y_test[col], y_pred[:, idx])
    return mae
```

Обучаем модели, считаем метрики. *_alpha - гиперпараметр, его можно поварьировать

```python
# Ridge
ridge_alpha = 1.0
ridge_model = ridge_regression(X_train, y_train, ridge_alpha)
ridge_mae = evaluate_model(ridge_model, X_test, y_test)
ridge_mae
```

```python
# Lasso
lasso_alpha = 1.0
lasso_model = lasso_regression(X_train, y_train, lasso_alpha)
lasso_mae = evaluate_model(lasso_model, X_test, y_test)
lasso_mae
```

```python
# Elastic Net
elastic_alpha = 0.1
elastic_l1_ratio = 0.1 # тоже гиперпараметр, тоже можно варьировать
elastic_net_model = elastic_net_regression(X_train, y_train, elastic_alpha, elastic_l1_ratio)
elastic_mae = evaluate_model(elastic_net_model, X_test, y_test)
elastic_mae
```

```python
# Polynomial
poly_degree = 3
poly_model = polynomial_regression(X_train, y_train, poly_degree)
poly_mae = evaluate_model(poly_model, X_test, y_test)
poly_mae
```
